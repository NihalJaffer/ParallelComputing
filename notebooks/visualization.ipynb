```python
import matplotlib.pyplot as plt
import numpy as np


algorithms = ['Right-Hand Rule', 'A* Search', 'Bidirectional']
moves = [1279, 127, 127]  # Number of moves for each algorithm
execution_times = [0.0017, 0.0022, 0.0008]  # Time taken in seconds
moves_per_second = [756951, 56916, 162600]  # Moves per second

# Create a figure with 2 subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Plot 1: Number of Moves Comparison
bars1 = ax1.bar(algorithms, moves, color=['blue', 'green', 'red'])
ax1.set_ylabel('Number of Moves')
ax1.set_title('Path Length Comparison')
ax1.set_ylim(0, max(moves) * 1.1)  # Add some space for labels
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# Add data labels on top of each bar
for bar in bars1:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, height + 10,
            f'{height}', ha='center', va='bottom')

# Plot 2: Execution Time Comparison
bars2 = ax2.bar(algorithms, execution_times, color=['blue', 'green', 'red'])
ax2.set_ylabel('Execution Time (seconds)')
ax2.set_title('Execution Time Comparison')
ax2.set_ylim(0, max(execution_times) * 1.2)  # Add some space for labels
ax2.grid(axis='y', linestyle='--', alpha=0.7)

# Add data labels on top of each bar
for bar in bars2:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2, height + 0.0001,
            f'{height:.6f}', ha='center', va='bottom')

plt.tight_layout()
plt.savefig('algorithm_comparison_1.png')
plt.show()

# Create another figure for moves per second comparison
plt.figure(figsize=(10, 6))
bars3 = plt.bar(algorithms, moves_per_second, color=['blue', 'green', 'red'])
plt.ylabel('Moves per Second')
plt.title('Processing Efficiency Comparison')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add data labels on top of each bar
for bar in bars3:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 10000,
            f'{height:,}', ha='center', va='bottom')

plt.tight_layout()
plt.savefig('algorithm_comparison_2.png')
plt.show()



```
